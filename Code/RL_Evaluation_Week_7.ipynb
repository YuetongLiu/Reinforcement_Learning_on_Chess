```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Evaluation
## Load Data 
Random, Q_learning and Policy 

```python
df = pd.read_csv('PG.csv')
pg = df['0']
df = pd.read_csv('Q.csv')
q = df['0']
df = pd.read_csv('Random.csv')
r = df['0']
len(r)
```

## Skip Sampling

In every 4 consecutive samples, draw one. Therefore all samples are approximately independent.

```python
skip = 4
pg_samples = []
q_samples = []
r_samples = []
for i in range(len(pg)):
        if i % skip == 0:
            pg_samples.append(pg[i])
            q_samples.append(q[i])
            r_samples.append(r[i])
len(r_samples)
```

## Density Plot

### random

```python
j = 0
bk = 200
for i in range(len(r_samples)):
    if (i+1) % bk == 0:
        plt.hist(r_samples[j:i], color = 'blue', edgecolor = 'black',
                 bins = int(180/5))
        plt.xlim(xmin=0, xmax = 200)
        plt.title('Iteration:'+str((j+1)*4)+'-'+str((i+1)*4))
        plt.show()
        j = i
```

### Q learning

```python
j = 0
bk = 200
for i in range(len(q_samples)):
    if (i+1) % bk == 0:
        plt.hist(q_samples[j:i], color = 'blue', edgecolor = 'black',
                 bins = int(180/5))
        plt.title('Iteration:'+str((j+1)*4)+'-'+str((i+1)*4))
        plt.show()
        j = i
```

### Policy Gradient

```python
j = 0
bk = 200
for i in range(len(pg_samples)):
    if (i+1) % bk == 0:
        plt.hist(pg_samples[j:i], color = 'blue', edgecolor = 'black',
                 bins = int(180/5))
        plt.title('Iteration:'+str((j+1)*4)+'-'+str((i+1)*4))
        plt.show()
        j = i
```
